```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rcParams
import matplotlib
```

```python
# NEW - import rcParams and set font settings
from matplotlib import rcParams

rcParams["font.family"] = "sans-serif"
rcParams["font.sans-serif"] = ["Arial"]
```

```python
ames_data = pd.read_csv('./data/ames.csv')
ames_data.head()
```

```python
from skimpy import clean_columns

ames_data = clean_columns(ames_data)
ames_data.columns.tolist()
ames_data

```

```python
ames_data.columns = ames_data.columns.str.replace(pat=" ", repl="_")
ames_data.columns = ames_data.columns.str.lower()
ames_data
```

```python
print(ames_data['central_air'].unique())
```

```python
figure, axes = plt.subplots(figsize=(6,5))
axes.scatter(x=ames_data['sale_price'], y=ames_data['gr_liv_area']);
```

```python
fig, (axes1,axes2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))  # Adjust the size and layout as needed
plt.tight_layout()  # Adjusts subplot params
fig.subplots_adjust(hspace=0.2, wspace=0.2)  # Adjust horizontal and vertical spaces

axes1.scatter(x=ames_data['sale_price'], y=ames_data['gr_liv_area'])

air_conditioning = ames_data["central_air"].unique().tolist() # Get a list with the unique values of the central_air column.
air_conditioning.sort() # Sorting this list to alphabetical order

for air_con_present in air_conditioning:
    # NEW - Get the subset of our data corresponding to the presence of air conditioning or not.
    air_con_rows = ames_data[ames_data["central_air"] == air_con_present]
    
    axes2.scatter(x=air_con_rows["sale_price"], 
                 y=air_con_rows["gr_liv_area"],
                 label=air_con_present) # NEW -  Gives each value a label for our legend

plt.legend();
```
