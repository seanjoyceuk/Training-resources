```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
coin = ['head', 'tail']
```

```python
np.random.seed(123456789)
coin_flip = np.random.choice(coin,size=100)
```

```python
count = np.unique(coin_flip, return_counts=True)
count
```

```python
plt.bar(count[0],count[1])
plt.title('Coin Tosses')
plt.xlabel('Outcomes')
plt.ylabel('Frequency')
```

```python
# toss biased coin

probability = [0.7,0.3]
np.random.seed(123456789)
coin_flip_biased = np.random.choice(coin,p=probability, size=30)
coin_flip_biased
```

```python
new_count = np.unique(coin_flip_biased, return_counts=True)
new_count
```

```python
probability = [0.7,0.3]
np.random.seed(123456789)
coin_flip_biased = np.random.choice(coin,p=probability, size=30)
coin_flip_biased
new_count = np.unique(coin_flip_biased, return_counts=True)
plt.bar(new_count[0], new_count[1])
plt.title('Count of results of biased coin toss')
plt.xlabel('Outcome')
plt.ylabel('Count')
```

```python
# Die roll

die = np.arange(1,7)
die_rolled = np.random.choice(die,10)
count_number = np.unique(die_rolled, return_counts=True)
plt.bar(count_number[0], count_number[1])
plt.title('Count of results of die rolls')
plt.xlabel('Outcome')
plt.ylabel('Count')
```

```python
# Biased dice roll

probability = np.array([2,1,1,1,1,1])/7

outcome = np.random.choice(die, p=probability, size=100)

next_count = np.unique(outcome, return_counts=True)

plt.bar(next_count[0], next_count[1])
plt.title('Count of results of biased die roll')
plt.xlabel('Outcome')
plt.ylabel('Count')
```

```python
# Two dice roll

two_dice = np.arange(2,13)

two_prob = np.array([1,2,3,4,5,6,5,4,3,2,1])/36

two_dice_roll = np.random.choice(two_dice, size = 100000000, p=two_prob)

count_number = np.unique(two_dice_roll, return_counts=True)
plt.bar(count_number[0], count_number[1])
plt.title('Count of results of two die rolls')
plt.xlabel('Outcome')
plt.ylabel('Count')
```
