```python
mushroom_data = data.copy()
```

```python
# Change - to _ for headers
mushroom_data.columns=mushroom_data.columns.str.replace('-','_')
mushroom_data
```

```python
mushroom_data.columns
```

```python
mushroom_data.isna().sum()
```

```python
mushroom_data['class'].unique()
```

```python
mushroom_data.describe(include='all')
```

```python
mushroom_data['cap_diameter'].mean()
```

```python
# Overwrite existing class column with more detail drawn from the text file

classes = dict(p = 'poisonous', e = 'edible')
mushroom_data['class'] = mushroom_data['class'].map(classes)
mushroom_data
```

```python
# Different way of changing the values using the data dictionary

mushroom_data['season'] = mushroom_data['season'].map(dict(s = 'spring', 
                                                    u = 'summer', 
                                                    a = 'autumn', 
                                                    w = 'winter'))
mushroom_data['season']
```

```python
mushroom_data['class'].value_counts()
```

```python
# Visualisation

mushroom_data['class'].value_counts().plot(kind='bar')
plt.show()
```

```python
mushroom_data['stem_height'].plot(kind='hist', bins=100)
plt.title('Stem Height Distribution')
plt.xlabel('Stem Height')
plt.show()
```

```python
# Bivariate visualation

plt.scatter(x=mushroom_data['stem_height'],y=mushroom_data['stem_width'])
plt.title('Stem Height vs Width')
plt.xlabel('Stem Height')
plt.ylabel('Stem Width')
plt.show()
```

```python
# 2 x categorical catergories - class and color
classes = dict(n='brown', b='buff', g='grey', r='green', p='pink',
								u='purple', e='red', w='white', y='yellow', l='blue', 
								o='orange',  k='black')
mushroom_data['cap_color'] = mushroom_data['cap_color'].map(classes)

ct = pd.crosstab(mushroom_data['class'], mushroom_data['cap_color'])
ct
```
