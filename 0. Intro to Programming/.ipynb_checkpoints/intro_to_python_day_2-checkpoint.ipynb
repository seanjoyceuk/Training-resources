## Key Learning

- Can run cell using Ctrl + Enter or Shft + Enter
- Get detail on data using .info() and .describe()
- Get data for objects too using .describe(include='all')
- Use describe function to get specific data on specific data types (include/exclude='float')
- Finding specific values using .describe() eg. mean, median, std
- Calculating the range manually
- Calculating square roots using math library
- Identifying data types using type()
- Finding missing values using .isna() or .isnull()
- Finding the sum of missing values using .isna().sum() or .isnull().sum()
- Imputing missing values with the mean or a specific number
- Using inplace=True to write the values to the DF, rather than having to re-assign to a variable
- Dropping a whole column using .drop('Column', axis = 1, inplace=True)
- Dropping rows with missing values using .dropna()
- Using .groupby as a filtering tool that carries out aggregation
- Using pandas to do basic visualisations
- Using some basic matplotlib functions - xlabel, ylabel, title



```python
import pandas as pd
```

```python
# Use .copy() at this stage as best practice and whenever you create a new subset of data

ames_data = pd.read_csv('ames.csv')
```

```python
ames_data.head(10)
```

```python
ames_data.info()
```

```python
ames_data.describe()
```

```python
# Get data for objects too
# Unique gives you the number of unique values
# top = highest value
# freq = value of the top value

ames_data.describe(include='all')
```

```python
# Only objects
ames_data.describe(include='object')
```

```python
# Only float
ames_data.describe(include='float')
```

```python
# Exclude float
ames_data.describe(exclude='float')
```

```python
# Exclude multiple data types

ames_data.describe(exclude=['float','object'])
```

```python
# Finding specific values

ames_data['MS SubClass'].mean()
```

```python
# Round to 2 digits - don't have to specvify this parameter, but it is best practice to define the parameters
# Shit Tabx3 gives you lots of info

round(ames_data['MS SubClass'].mean(), ndigits=2)
```

```python
ames_data['MS SubClass'].median()
```

```python
ames_data['MS SubClass'].std()
```

```python
# To create range, need to manually create a function

ames_data['MS SubClass'].max() - ames_data['MS SubClass'].min()
```

```python
# Doing a square root

import math
math.sqrt(ames_data['MS SubClass'].var())
```

```python
# To avoid having to specify math. each time

from math import sqrt

sqrt(ames_data['MS SubClass'].var())
```

```python
# This operates as a pandas object, so if you apply methods, it will call from pandas

type(ames_data)
```

```python
# Imputation - Filling in values with calculated values (can use ML)
# Deletion - Getting rid of values
```

```python
# Finding missing values -  can use .isna or .isnull

ames_data.isna()
```

```python
# Sum of null values

ames_data.isnull().sum()
```

```python
missing_value = ames_data.isnull().sum()
filter_columns = missing_value[(missing_value>=1)].sort_values(ascending = False)
filter_columns
```

```python
# Always request a data dictionary - key for features
```

```python
# Calculate the mean for imputation
mean_value_MVA = round(ames_data['Mas Vnr Area'].mean(), ndigits=2)
mean_value_MVA
```

```python
# Replace the values in missing data
# inplace=True - Applies it to your dataset
# Inplace=True is the same as re-assigning the data to the same variable

ames_data['Mas Vnr Area'].fillna(mean_value_MVA, inplace=True)
```

```python
# Missing values have gone

ames_data.info()
```

```python
# Check missing values in a specific column

ames_data['Mas Vnr Area'].isna().sum()
```

```python
# Imputing a specific numerical value

ames_data['Bsmt Half Bath'].fillna(-10, inplace=True)
```

```python
ames_data['Bsmt Half Bath'].isnull().sum()
```

```python
# Dropping whole column, asw there are hardly any values for this column

# Axis = 1 refers to droppoing from a column, rather than a row

ames_data.drop('Pool QC', axis = 1, inplace=True)
```

```python
# Dropping rows with missing values

ames_data.dropna()
```

```python
# Using groupby as a filtering tool that carries out aggregation - this calculates average sale price for each neighbourhood

mean_price_by_neighbourhood = ames_data.groupby('Neighborhood')['SalePrice'].mean().sort_values(ascending=True)
mean_price_by_neighbourhood
```

```python
# Using pandas to do visualisations
# barh presents it horizontally

mean_price_by_neighbourhood.plot(kind='barh')
```

```python
import matplotlib.pyplot as plt
```

```python
mean_price_by_neighbourhood.plot(kind='barh', figsize=(20,30))
plt.xlabel('Neighborhood')
plt.ylabel('Average Price')
plt.title('Average house price by neighbourhood)')
```
